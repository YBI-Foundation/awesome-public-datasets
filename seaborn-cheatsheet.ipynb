# **Seaborn CheatSheet**

```python
# **Seaborn**

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt

# [Inbuid Dataset](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)

# inbuild datasets
sns.get_dataset_names()

# load inbuild data
iris = sns.load_dataset('iris')

iris.head()

# [Line Plot](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

# lineplot
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)

#set the title using Matplotlib
fig,ax = plt.subplots()
ax.set_title('Create Title with Matplotlib')

# lineplot
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)

plt.show()

# changing the figure size
fig,ax = plt.subplots(figsize = (10, 5))
ax.set_title('Create Title with Matplotlib')

# lineplot
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)

plt.show()

# [Relational Plot](https://seaborn.pydata.org/generated/seaborn.relplot.html)

# relplot
sns.relplot(x='sepal_width', y='species', data=iris)

# [Scatter Plot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

# scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)

# [Bar Plot](https://seaborn.pydata.org/generated/seaborn.barplot.html)

# bar plot
sns.barplot(x='species', y='sepal_length', data=iris)

# [Count Plot](https://seaborn.pydata.org/generated/seaborn.countplot.html)

# count plot
sns.countplot(x='species', data=iris)

# [Box Plot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

# box plot
sns.boxplot(x='species', y='sepal_width', data=iris)

# [Violin Plot](https://seaborn.pydata.org/generated/seaborn.violinplot.html)

# violin plot
sns.violinplot(x='species', y='sepal_width', data=iris)

# [Strip Plot](https://seaborn.pydata.org/generated/seaborn.stripplot.html)

# strip plot
sns.stripplot(x='species', y='sepal_width', data=iris)

# [Swarm Plot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)

# swarm plot
sns.swarmplot(x='species', y='sepal_width', data=iris)

# [Cat Plot](https://seaborn.pydata.org/generated/seaborn.catplot.html)

# cat plot
sns.catplot(x='species', y='sepal_width', data=iris)

# [Histogram](https://seaborn.pydata.org/generated/seaborn.histplot.html)

# histogram
sns.histplot(x='species', y='sepal_width', data=iris)

# [Dis Plot](https://seaborn.pydata.org/generated/seaborn.displot.html#)

# dis plot - univariate
sns.displot(iris['sepal_width'])

# dis plot - bivariate
sns.displot(x='sepal_length', y='sepal_width', data=iris)

# [Join Plot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

# join plot
sns.jointplot(x='species', y='sepal_width', data=iris)

# [Pair Plot](https://seaborn.pydata.org/generated/seaborn.pairplot.htmlhttps://)

# pair plot
sns.pairplot(data=iris, hue='species')

# [Rug Plot](https://seaborn.pydata.org/generated/seaborn.rugplot.html)

# rug plot
sns.rugplot(data=iris)

# [Kernel Density Plot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html)

# kde plot
sns.kdeplot(x='sepal_length', y='sepal_width', data=iris)

# [ECDF Plot](https://seaborn.pydata.org/generated/seaborn.ecdfplot.html)

# ecdf plot
sns.ecdfplot(x='sepal_length', data=iris)

# [Reg Plot](https://seaborn.pydata.org/generated/seaborn.regplot.html)

# reg plot
sns.regplot(x='sepal_length', y='sepal_width', data=iris)

# [LM Plot](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

# lm plot regression and logistic = True
sns.lmplot(x='sepal_length', y='sepal_width', data=iris)

# [Heat Plot](https://seaborn.pydata.org/generated/seaborn.heatmap.html)

# heatmap
sns.heatmap(iris.corr())

# [Cluster Map](https://seaborn.pydata.org/generated/seaborn.clustermap.html)

# cluster map
sns.clustermap(iris.corr())
```
